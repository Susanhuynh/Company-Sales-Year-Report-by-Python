```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company_sale = pd.read_csv("OneDrive/Machine Learning/Dataset/company_sales_data.csv")
```

```python
company_sale.head()
```

```python
## Company profit for each month
plt.figure(figsize=(15,6))
plt.plot(company_sale["month_number"], company_sale["total_profit"],
         label="Company profit for each month", color="r", linestyle="--", linewidth=3,marker="o", markerfacecolor="k", markersize = 10)
plt.xlabel("Month Number")
plt.ylabel("Profit")
plt.xticks(company_sale["month_number"].tolist())
plt.yticks([100000,200000,300000,400000,500000])
plt.legend(loc=4)
```

```python
## Sales of all products
plt.figure(figsize=(15,6))
plt.plot(company_sale["month_number"], company_sale["facecream"], marker="o", label="Facecream Sales Data")
plt.plot(company_sale["month_number"], company_sale["facewash"], marker="o", label="Facewash Sales Data")
plt.plot(company_sale["month_number"], company_sale["toothpaste"], marker="o", label="Toothpaste Sales Data")
plt.plot(company_sale["month_number"], company_sale["bathingsoap"], marker="o", label="Bathingsoap Sales Data")
plt.plot(company_sale["month_number"], company_sale["shampoo"], marker="o", label="Shampoo Sales Data")
plt.plot(company_sale["month_number"], company_sale["moisturizer"], marker="o", label="Moisturizer Sales Data")
plt.xticks(company_sale["month_number"].tolist())
plt.legend(loc="upper left")
```

```python
## Toothpaste for each month
plt.figure(figsize=(12,6))
plt.scatter(company_sale["month_number"], company_sale["toothpaste"], label="Tooth paste Sales Data")
plt.xticks(company_sale["month_number"].tolist())
plt.xlabel("Month Number")
plt.ylabel("Nunber of Units Sold")
plt.legend(loc="upper left")
plt.grid(linestyle="--")
```

```python
## FACECREAM and FACEWASH 
plt.figure(figsize=(12,6))
x=company_sale["month_number"].tolist()
width = 0.35
facecream = plt.bar([a-width/2 for a in x], company_sale["facecream"], width, label="Facecream Sales Data")
facecream = plt.bar([a+width/2 for a in x], company_sale["facewash"], width, label="Facewash Sales Data")
plt.xticks(x)
plt.legend(loc="upper left")
plt.xlabel("Month Number")
plt.ylabel("Sales Units in number")
plt.grid(linestyle="--")
plt.title("Facewash and facecream sales data")
```

```python
## BATHING SOAT PER MONTHS
plt.figure(figsize=(12,6))
plt.bar(company_sale["month_number"], company_sale["bathingsoap"])
plt.xticks(company_sale["month_number"].tolist())
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.title("bathing soap sales data")
plt.grid(linestyle="--")
```

```python
total_sales = company_sale["facecream"].sum() + company_sale["facewash"].sum() + company_sale["toothpaste"].sum() + company_sale["bathingsoap"].sum() + company_sale["moisturizer"].sum() + company_sale["shampoo"].sum()
```

```python
plt.figure(figsize=(12,6))
monthList = company_sale["month_number"].tolist()
productList = company_sale.columns[1:7].tolist()
size=[]
for a in productList:
    total = company_sale[a].sum()
    per = (total/total_sales)*100
    size.append(per)
plt.pie(size, labels=productList, autopct='%1.1f%%')
plt.legend(loc="lower right")
plt.title("Sales Data")
```

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(company_sale["month_number"], company_sale["bathingsoap"], marker="o", color = "k", linewidth=3)
ax[0].set_title("Sales data of bathing soap")
ax[1].plot(company_sale["month_number"], company_sale["facewash"], marker="o", color="r", linewidth=3)
plt.ylabel("Sales number in number")
plt.xticks(company_sale["month_number"].tolist())
plt.xlabel("Month Number")
```
